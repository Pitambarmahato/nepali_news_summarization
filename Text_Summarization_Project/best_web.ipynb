```python
def get_data(url, l, h):
    data = requests.get(url)
    soup= BeautifulSoup(data.text, 'html.parser')
    lst_url = []
    for content in soup.select('div a', class_ = 'row bishesh news-cat-list alt'):
        url = content.get('href')
        lst_url.append(url)
    if 'javascript:void(0)' in lst_url:
        lst_url.remove('javascript:void(0)')
#     print(lst_url)
    lst_url1 = lst_url[l:h]
    content = []
    for i in lst_url1:
        cont = requests.get(i)
        soup = BeautifulSoup(cont.text, 'html.parser')
        for head in soup.find_all('span', class_= 'news-big-title'):
            content.append(head.text)
    body = []
    for i in lst_url1:
        bod = requests.get(i)
        soup = BeautifulSoup(bod.text, 'html.parser')
    #     print(soup.prettify())
        body1 = []
        for main in soup.select('div p', class_ = 'detail-box col-md-12'):
    #         print(''.join(main.text))
            body1.append(main.text)
        body2 = ''.join(body1)
        body.append(body2)

    # print(body)
    d = {'Title': content, 'Body': body}
    return d
```

```python
data = get_data('https://www.setopati.com/sports', 35, 80)
```
